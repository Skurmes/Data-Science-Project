```python
#import library
import numpy as np
import pandas as pd
```

```python
#store the data
df = pd.read_csv('DOGE-USD.csv')
df = df.set_index(pd.DatetimeIndex(df['Date'].values)) #setting the index to be the date
df #showing the data
```

```python
df = df[['Close']] #getting the close price and showing it
df
```

```python
prediction_days = 1 #variables to store the number of days that we want to predict
df['Prediction'] = df[['Close']].shift(-predection_days) #new coloum and namaning it prediction.
df
```

```python
#creating the independent data set (x)
X = np.array(df.drop(['Prediction'], 1))
X = X[:len(df)-prediction_days -1]
print(X)
```

```python
#creating a dependent data set (y)
y = np.array(df['Prediction'])
y = y[:- prediction_days - 1]
print(y)
```

```python
#spliting the data into traning and testing data set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (X, y, test_size = 0.2)
```

```python
#using RandomForestRegressor for the model
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators = 2, random_state = 587)
forest.fit(x_train, y_train)
print(forest.score(x_test, y_test))
```

```python
#show how close the predectied values and the actual value
predicition = forest.predict(x_test)
print(predicition)
print()
print(y_test)
```

```python
#getting the validation data
#creating a variable to store all the rows except the last n rows
temp_df = df[:-prediction_days]
x_val = temp_df.tail(1)['Close'][0] #creating a variable to store the independent price value
print(x_val) #showing the data
```

```python
prediction = forest.predict([[x_val]])
print('The price of Dogecoin in' , prediction_days, 'days is predicted to be', prediction) #printing the price of Dogecoin for the next n days
print('The actual price was', temp_df.tail(1)['Prediction'][0])#printing the actual value for the next n days
```
